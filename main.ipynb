```python
from src.pseudo_hilbert_curve import PseudoHilbertCurve
from src.snake_curve import SnakeCurve
from src.curve_utils.dimension import Dimension
from src.curve_utils.plot_utils import plot_curve
from matplotlib import pyplot as plt
from src.curve_utils.plot_utils import get_xy_from_curve
```

```python
dims = [
    Dimension(2, 2),
    Dimension(4, 4),
    Dimension(8, 8),
]

num_els = [
    2 * 2,
    4 * 4,
    8 * 8,
]

curves = [
    PseudoHilbertCurve(num_els[0], dims[0], 2, False),
    PseudoHilbertCurve(num_els[1], dims[1], 2, False),
    PseudoHilbertCurve(num_els[2], dims[2], 2, False),
]

fig, ax = plt.subplots(1, 3, figsize=(6, 2))
for i, curve in enumerate(curves):
    x, y = get_xy_from_curve(curve)
    ax[i].set_xticks(range(int(min(x)), int(max(x)) + 1))
    ax[i].set_yticks(range(int(min(y)), int(max(y)) + 1))
    ax[i].plot(x, y)
plt.tight_layout()
```
